```python
# Library Import
from google_play_scraper import reviews_all, reviews, Sort
import pandas as pd
```

```python
# Get 30K reviews app of shopee
scrapreview = reviews(
    'com.shopee.id',
    lang='id',
    country='id',
    sort=Sort.MOST_RELEVANT,
    count = 30000,
    filter_score_with=5
)
```

```python
# Save the scraping result in to csv
data = pd.DataFrame(scrapreview[0])
data.to_csv('review_shopee.csv', index=False)
```
